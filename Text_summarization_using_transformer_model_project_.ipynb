```python
#step 2
# import dataset
from datasets import load_dataset

# Load CNN/DailyMail dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")

#You can also use 'train', 'validation' or 'test' splits
#sample article
article = dataset["test"][0]["article"]
print(article)
```

```python
# checking dataset structure
print(dataset)
print(dataset["test"])
print(len(dataset["test"]))
```

```python

# Step 3: Define your cleaning function
def clean_text(text):
    if text:  # Check if text is not None
        text = text.strip()  # Remove leading/trailing spaces
        text = text.replace("\n", " ")  # Replace newlines with space
    return text

#  Apply cleaning to all articles
dataset = dataset.map(lambda x: {"article": clean_text(x["article"])})

#  Select an article
article_index = 1
article = dataset["test"][article_index]["article"]

# Check whether the article is empty or not
if article and article.strip():
    print(f"Selected Article (after cleaning): {article[:500]}...")
else:
    print("Article is empty after cleaning.")
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Load pre-trained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)
```

```python
#tokenizing entire dataset

# Define the preprocessing function
def preprocess_function(examples):
    inputs = tokenizer(
        examples["article"],
        truncation=True,
        padding="max_length",
        max_length=384
    )
    targets = tokenizer(
        examples["highlights"],
        truncation=True,
        padding="max_length",
        max_length=96
    )
    inputs["labels"] = targets["input_ids"]  # Set labels for training
    return inputs

tokenized_datasets = dataset.map(
    preprocess_function,
    batched=True,
    num_proc=4,  # Parallel processing (reduce if slow)
    batch_size=1000  # Prevent memory issues
)

```

```python
# Assign tokenized datasets for training, validation, and testing
train_dataset = tokenized_datasets["train"]
val_dataset = tokenized_datasets["validation"]
test_dataset = tokenized_datasets["test"]
```

```python

# Generate summary for a sample test article

# Select an article from the test set
article = dataset["test"][0]["article"]

# Tokenize the selected article
tokenized_input = tokenizer(article, return_tensors="pt", truncation=True, max_length=1024)
# Generate a summary
summary_ids = model.generate(tokenized_input.input_ids, max_length=150, num_beams=4, early_stopping=True)

# Decode the generated summary
generated_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Print output
print("\nGenerated Summary:", generated_summary)
print("\nReference Summary:", dataset["test"][0]["highlights"])
```

```python
from transformers import TrainingArguments, Trainer
from transformers import DataCollatorForSeq2Seq

from transformers import BartForConditionalGeneration, BartTokenizer

# Select a smaller dataset for training & validation
small_train_dataset = dataset["train"].select(range(5000))  # Use only 5000 samples
small_val_dataset = dataset["validation"].select(range(1000))  # Use 1000 samples for validation

# Tokenize the small datasets
tokenized_small_train_dataset = small_train_dataset.map(
    preprocess_function,
    batched=True,
    #remove_columns=small_train_dataset.column_names # remove this parameter
)

tokenized_small_val_dataset = small_val_dataset.map(
    preprocess_function,
    batched=True,
    #remove_columns=small_val_dataset.column_names # remove this parameter
)

# Define Training Arguments
training_args = TrainingArguments(
    output_dir="./bart_summarization",
    eval_strategy="epoch",  # Changed parameter name
    save_strategy="epoch",
    per_device_train_batch_size=2,  # Increased batch size if GPU allows
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=4,  # Helps when batch size is small
    learning_rate=5e-5,
    weight_decay=0.01,
    num_train_epochs=1,
    logging_dir="./logs",
    logging_steps=500,
    save_total_limit=2,
    fp16=True,  # Mixed precision for faster training on GPUs
    report_to="none"  # Disables Weights & Biases logging
)
# Initialize Data Collator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_small_train_dataset, # Use tokenized dataset here
    eval_dataset=tokenized_small_val_dataset,  # Use tokenized dataset here
    tokenizer=tokenizer,
    data_collator=data_collator
)


# Start Fine-Tuning
trainer.train()
```

```python
from rouge_score import rouge_scorer
import torch

# Initialize ROUGE scorer
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

# Step 3: Select an article from the cleaned dataset
article_index = 0  # Change this index to select different articles
article = dataset["test"][article_index]["article"]
reference_summary =dataset["test"][article_index]["highlights"]  # Ground truth summary

# Step 4: Print to verify
print("Selected Article (Cleaned):", article[:500])
print("Reference Summary (Cleaned):", reference_summary)

# Tokenize and generate a summary
tokenized_input = tokenizer(article, return_tensors="pt", truncation=True, max_length=1024)
# Move input tensors to the same device as the model
tokenized_input.input_ids = tokenized_input.input_ids.to(model.device)

summary_ids = model.generate(tokenized_input.input_ids, max_length=150, num_beams=4, early_stopping=True)
generated_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Compute ROUGE scores
rouge_scores = scorer.score(reference_summary, generated_summary)

# Print ROUGE scores
print(f"Generated Summary: {generated_summary}\n")
print(f"Reference Summary: {reference_summary}\n")

print("ROUGE Scores:")
print(f"ROUGE-1: {rouge_scores['rouge1'].fmeasure:.4f}")
print(f"ROUGE-2: {rouge_scores['rouge2'].fmeasure:.4f}")
print(f"ROUGE-L: {rouge_scores['rougeL'].fmeasure:.4f}")
```
